# Object Oriented Programming (init, classes, setter, and getter)

```python
class EmployeePersonalInfo(object): #BIG O NOTATION: Exponential Complexity (O2^(n)) # Class for personal infomation to include names generated from names libary, and addresses generated from faker 

  def __init__(self, first_name, last_name, address): # Init is the constructor in object oriented programming, this one includes names and addresses, it intializes the attributes 
      
    self.first_name = first_name                      # Variables to be used for later 
    self.last_name = last_name
    self.address= address  
  
  def set_first_name(self, first_name):  # Setter for first name 

    self.first_name = first_name

  def get_first_name(self):
    return self.first_name              # Return function for Setter 

  def set_last_name(self, last_name):

    self.last_name = last_name

  def get_last_name(self):              # Getter for names, the same procedure like stated before, however now it is getter 

    return self.last_name

  def set_address(self, address):       # Same procdures as before 
 
    self.address  = address

  def get_address(self):

    return self.address

  def __str__(self):

    return f" Employee name is:{self.get_first_name()} {self.get_last_name()}, Employee address is: {self.get_address()}" # Printing to get addresses and names, these strings will appear on the function that produces the list 

class EmployeeNumericInfo(object):  # Now we do numeric info, such as salary, to be sorted at a later time  


  def __init__(self, salary,IDNUMBER): # Init except for a different list, this time for numeric items within for a employee database 
      
    self.salary = salary
    self.IDNUMBER = IDNUMBER
  
  def set_salary(self, salary):

    self.salary = salary

  def get_salary(self):

    return self.salary

  def set_IDNUMBER(self, IDNUMBER):

    self.IDNUMBER= IDNUMBER

  def get_IDNUMBER(self):

    return self.IDNUMBER

  def __lt__(self, other):                     # Necessary function for sorting at a later time 

     if self.salary == other.salary:
      return self.IDNUMBER < other.IDNUMBER
     return self.salary < other.salary 

  def __str__(self):

    return f" Employee salary is:{self.get_salary()}, Employee ID Number is: {self.get_IDNUMBER()}" # Now we do numeric info, such as age, to be sorted at a later time  

class EmployeeAgeandGender(object):          # The same procedure as before, but with a diferent class 

  def __init__(self, age):
      
    self.age = age
  
  def set_age(self, age):

    self.age = age

  def get_age (self):
    return self.age

  def __lt__(self, other):

      return self.age < other.age

  def __str__(self):

    return f" Employee age is:{self.get_age()}"
```

```python
def python_search(L, e): #BIG O NOTATION: Linear Complexity (O(n)) # Python search to identify numbers for later 
  print('List length:', len(L))
  print('Size:', sys.getsizeof(L))
  print('Searching for', e)
  t1 = time.time()
  if e in L:
    print('Found it')
  else:
    print('Does not exist in the list')
  t2 = time.time()
  print('Time consumed:', t2 - t1)
```
